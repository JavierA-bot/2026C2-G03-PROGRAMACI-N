# Problema #1: Calculadora de propina inteligente

# Problema #3: Clasificador de edad

## Problema

Realizar un clasificar de edad a partir del dato ingresado

## Entrada
#Usuario debe ingresar la edad

edad = int(input("Ingresa tu edad: "))

## Proceso
#Convertir la entrada a int, luego Evaluar la edad:

# Problema #4: Validador de compra con descuento

## Problema

Realizar un validador de compra con descuento

## Entrada
#Solicitar nombre del cliente, monto de compra, si es cliente frecuente

cliente = input("Ingresa el nombre del cliente: ")
monto = float(input("Ingresa el monto de la compra: "))
frecuente = input("¿Es cliente frecuente? (si/no): ")

## Proceso
#Convertir el monto a float
#Evalua si el cliente es frecuente: Si es frecuente → aplicar 10% de descuento, Si no → no hay descuento
#Calculo: descuento = monto * 0.10 (si aplica), total final = monto - descuento
#Usar operadores lógicos and/or (de forma simple)

if frecuente == "si" or frecuente == "sí":
    descuento = monto * 0.10
else:
    descuento = 0

total_final = monto - descuento 

## Salida esperada
#Mostrar el nombre del cliente, descuento y total final a pagar

print(f"Cliente: {cliente}")
print(f"Descuento aplicado: {descuento}")
print(f"Total a pagar: {total_final}")

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | | | |
| Caso límite | | | |
| Caso inválido | | | |


```python
cliente = input("Ingresa el nombre del cliente: ")
monto = float(input("Ingresa el monto de la compra: "))
frecuente = input("¿Es cliente frecuente? (si/no): ")

if frecuente == "si" or frecuente == "sí":
    descuento = monto * 0.10
else:
    descuento = 0

total_final = monto - descuento

print(f"Cliente: {cliente}")
print(f"Descuento aplicado: {descuento}")
print(f"Total a pagar: {total_final}")
```
